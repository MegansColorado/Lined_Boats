## Now lets try to find the edges and featurize them...

skimage.filters.sobel_h(image[, mask]) - Find the horizontal edges of an image using the Sobel transform.

skimage.filters.sobel_v(image[, mask]) - Find the vertical edges of an image using the Sobel transform.

```python
#create sobel features

sobel_h = []
sobel_v = []

for i in range(3):
    sobel_h.append(filters.sobel_h(sketch_rgb[:,:,i]))
    sobel_v.append(filters.sobel_v(sketch_rgb[:,:,i]))
    

```

```python
#NOW Create a data frame with all of the features for that image

x = []
y = []

red = []
green = []
blue = []

r_hzntl = []
g_hzntl = []
b_hzntl = []

r_vert = []
g_vert = []
b_vert = []


for i in range(sketch_rgb.shape[0]):
    for j in range(sketch_rgb.shape[1]):
        x.append(j)
        y.append(i)
        
        red.append(sketch_rgb[i,j,0])
        green.append(sketch_rgb[i,j,1])
        blue.append(sketch_rgb[i,j,2])
        
        r_hzntl.append(sobel_h[0][i,j])
        g_hzntl.append(sobel_h[1][i,j])
        b_hzntl.append(sobel_h[2][i,j])

        r_vert.append(sobel_v[0][i,j])
        g_vert.append(sobel_v[1][i,j])
        b_vert.append(sobel_v[2][i,j])
        #sobel.append(sobel[i,j])
df = pd.DataFrame({
'x':x, 
'y':y,
'red':red,
'green':green,
'blue':blue,
'r_hzntl': r_hzntl,
'g_hzntl': g_hzntl,
'b_hzntl': b_hzntl,
'r_vert': r_vert,
'g_vert': g_vert,
'b_vert': b_vert })
df
```

```python
#PCA to remove features, first a scree plot
pca = PCA(whiten=True) 
pca_image = pca.fit_transform(df)

```
